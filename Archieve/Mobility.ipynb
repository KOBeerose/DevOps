```python
import pandas as pd

s1 = pd.read_csv("Synthèse S1 DATA après rattrapage.csv")
```

```python
s1.rename( columns={'Unnamed: 5':'name'}, inplace=True )
s1.rename( columns={'Unnamed: 40':'mark_s1'}, inplace=True )
```

```python
s1 = s1[["name", "mark_s1"]]
```

```python
s1.dropna(inplace=True)
s1 = s1[s1.mark_s1 != "Moy\nS1"]
s1 = s1.reset_index(drop=True)

#s1.drop([2,35],inplace=True)
```

```python
s1
```

```python
s1.mark_s1 = s1.mark_s1.astype(float)
```

```python
s1["mark_s1"] = s1["mark_s1"].apply(lambda x: x/100)
```

```python
s1
```
